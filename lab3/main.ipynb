# Лабораторная работа: NumPy операции с массивами

## Задача 1: Операции с матрицами

### 1.1 Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt

# Настройка отображения графиков
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
```

### 1.2 Генерация случайной матрицы 8×8

```python
# Создаем случайную матрицу 8x8 с числами от -10 до 10
np.random.seed(42)  # Для воспроизводимости результатов
matrix = np.random.randint(-10, 11, (8, 8))

print("1. Исходная матрица 8x8:")
print(matrix)
print(f"\nРазмер матрицы: {matrix.shape}")
print(f"Тип данных: {matrix.dtype}")
```

### 1.3 Вывод центральной части матрицы 4×4

```python
# Извлекаем центральную часть 4x4
center_start = 2  # Начинаем с 3-й строки (индекс 2)
center_end = 6    # Заканчиваем 6-й строкой (индекс 5)
center_matrix = matrix[center_start:center_end, center_start:center_end]

print("2. Центральная часть матрицы 4x4:")
print(center_matrix)
print(f"\nРазмер центральной части: {center_matrix.shape}")

# Визуализация выделенной центральной части
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Исходная матрица с выделением центра
im1 = ax1.imshow(matrix, cmap='coolwarm', aspect='auto')
ax1.set_title('Исходная матрица 8x8')
ax1.set_xlabel('Столбцы')
ax1.set_ylabel('Строки')
plt.colorbar(im1, ax=ax1)

# Прямоугольник, показывающий выделенную область
import matplotlib.patches as patches
rect = patches.Rectangle((center_start-0.5, center_start-0.5), 
                         center_end-center_start, 
                         center_end-center_start,
                         linewidth=2, edgecolor='yellow', facecolor='none')
ax1.add_patch(rect)

# Центральная часть
im2 = ax2.imshow(center_matrix, cmap='coolwarm', aspect='auto')
ax2.set_title('Центральная часть 4x4')
ax2.set_xlabel('Столбцы')
ax2.set_ylabel('Строки')
plt.colorbar(im2, ax=ax2)

plt.tight_layout()
plt.show()
```

### 1.4 Удаление строк с минимальным элементом

```python
# Находим минимальный элемент
min_element = np.min(matrix)
print(f"Минимальный элемент матрицы: {min_element}")

# Находим строки, содержащие минимальный элемент
rows_with_min = np.any(matrix == min_element, axis=1)
print(f"\nСтроки с минимальным элементом (True - содержит):")
print(rows_with_min)
print(f"Индексы строк с минимальным элементом: {np.where(rows_with_min)[0]}")

# Удаляем эти строки
matrix_without_min_rows = matrix[~rows_with_min]

print("\n3. Матрица после удаления строк с минимальным элементом:")
print(matrix_without_min_rows)
print(f"\nУдалено строк: {np.sum(rows_with_min)}")
print(f"Новый размер матрицы: {matrix_without_min_rows.shape}")
```

### 1.5 Вставка строки из минимального элемента перед первой строкой

```python
# Создаем строку из минимального элемента
min_row = np.full((1, matrix.shape[1]), min_element)

# Вставляем строку в начало матрицы
matrix_with_min_row = np.vstack([min_row, matrix])

print("4. Матрица со строкой из минимальных элементов в начале:")
print(matrix_with_min_row)
print(f"\nНовый размер матрицы: {matrix_with_min_row.shape}")
print(f"Первая строка (добавленная): {matrix_with_min_row[0]}")

# Визуализация измененной матрицы
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Исходная матрица
im1 = ax1.imshow(matrix, cmap='coolwarm', aspect='auto')
ax1.set_title('Исходная матрица')
ax1.set_xlabel('Столбцы')
ax1.set_ylabel('Строки')
plt.colorbar(im1, ax=ax1)

# Матрица с добавленной строкой
im2 = ax2.imshow(matrix_with_min_row, cmap='coolwarm', aspect='auto')
ax2.set_title('С добавленной строкой\nминимальных элементов')
ax2.set_xlabel('Столбцы')
ax2.set_ylabel('Строки')
plt.colorbar(im2, ax=ax2)

# Матрица с выделенной добавленной строкой
im3 = ax3.imshow(matrix_with_min_row, cmap='coolwarm', aspect='auto')
ax3.set_title('С выделенной добавленной строкой')
ax3.set_xlabel('Столбцы')
ax3.set_ylabel('Строки')
plt.colorbar(im3, ax=ax3)

# Добавляем прямоугольник вокруг первой строки
rect = patches.Rectangle((-0.5, -0.5), matrix_with_min_row.shape[1], 1, 
                         linewidth=3, edgecolor='yellow', facecolor='none')
ax3.add_patch(rect)

plt.tight_layout()
plt.show()
```

### 1.6 Вычисление суммы и среднего арифметического

```python
# Вычисляем сумму и среднее
total_sum = np.sum(matrix)
average = np.mean(matrix)

print("5. Статистика по исходной матрице:")
print(f"   Сумма всех элементов: {total_sum}")
print(f"   Среднее арифметическое: {average:.4f}")

# Дополнительная статистика
print(f"\nДополнительная статистика:")
print(f"   Минимальный элемент: {np.min(matrix)}")
print(f"   Максимальный элемент: {np.max(matrix)}")
print(f"   Стандартное отклонение: {np.std(matrix):.4f}")
print(f"   Медиана: {np.median(matrix):.2f}")
print(f"   Диапазон (размах): {np.ptp(matrix)}")

# Визуализация распределения элементов матрицы
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Гистограмма распределения
ax1.hist(matrix.flatten(), bins=21, edgecolor='black', alpha=0.7, color='skyblue')
ax1.axvline(average, color='red', linestyle='--', linewidth=2, 
            label=f'Среднее: {average:.2f}')
ax1.set_xlabel('Значение элемента')
ax1.set_ylabel('Количество')
ax1.set_title('Распределение элементов матрицы')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Статистика в виде столбчатой диаграммы
stats_values = [total_sum, average, np.min(matrix), np.max(matrix), np.std(matrix)]
stats_labels = ['Сумма', 'Среднее', 'Минимум', 'Максимум', 'Стд. отклонение']
colors = ['blue', 'green', 'red', 'orange', 'purple']
bars = ax2.bar(stats_labels, stats_values, color=colors)
ax2.set_ylabel('Значение')
ax2.set_title('Статистика матрицы')
ax2.tick_params(axis='x', rotation=45)

# Добавляем значения на столбцы
for bar, value in zip(bars, stats_values):
    ax2.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
             f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

## Задача 2: Операции с одномерными массивами

### 2.1 Генерация двух массивов

```python
# Генерируем два массива размерностью 10
np.random.seed(123)  # Для воспроизводимости
m1 = np.random.randint(1, 11, 10)
m2 = np.random.randint(1, 11, 10)

print("Сгенерированные массивы:")
print(f"m1: {m1}")
print(f"m2: {m2}")
print(f"\nДлина m1: {len(m1)}, Длина m2: {len(m2)}")

# Визуализация массивов
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(range(len(m1)), m1, color='blue', alpha=0.7)
ax1.set_xlabel('Индекс')
ax1.set_ylabel('Значение')
ax1.set_title('Массив m1')
ax1.set_xticks(range(len(m1)))
ax1.grid(True, alpha=0.3)

ax2.bar(range(len(m2)), m2, color='green', alpha=0.7)
ax2.set_xlabel('Индекс')
ax2.set_ylabel('Значение')
ax2.set_title('Массив m2')
ax2.set_xticks(range(len(m2)))
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 2.2 Создание массива m3 (симметрическая разность)

```python
# Находим элементы, которые есть только в m1 или только в m2
m3 = np.setxor1d(m1, m2)

print("1. Элементы, которые есть только в m1 или только в m2 (m3):")
print(f"m3: {m3}")
print(f"\nДлина m3: {len(m3)}")
print(f"Отсортированный m3: {np.sort(m3)}")

# Проверим результаты
print(f"\nПроверка:")
print(f"Элементы только в m1: {np.setdiff1d(m1, m2)}")
print(f"Элементы только в m2: {np.setdiff1d(m2, m1)}")

# Визуализация
fig, ax = plt.subplots(figsize=(10, 5))

indices = np.arange(len(m3))
bars = ax.bar(indices, m3, color='orange', alpha=0.7)
ax.set_xlabel('Индекс в m3')
ax.set_ylabel('Значение')
ax.set_title('Массив m3 (симметрическая разность m1 и m2)')
ax.set_xticks(indices)
ax.grid(True, alpha=0.3)

# Добавляем значения на столбцы
for bar, value in zip(bars, m3):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
             f'{value}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### 2.3 Замена в m1 кратных 2 или 3 на 1

```python
# Создаем копию m1 для модификации
m1_modified = m1.copy()

# Находим элементы, кратные 2 или 3
mask = (m1_modified % 2 == 0) | (m1_modified % 3 == 0)
print(f"Маска для замены (True - заменить на 1): {mask}")
print(f"Количество элементов для замены: {np.sum(mask)}")
print(f"Элементы для замены: {m1_modified[mask]}")

# Заменяем элементы, кратные 2 или 3, на 1
m1_modified[mask] = 1

print("\n2. Массив m1 после замены кратных 2 или 3 на 1:")
print(f"Исходный m1:     {m1}")
print(f"Модифицированный: {m1_modified}")

# Визуализация изменений
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Исходный массив
colors1 = ['red' if mask[i] else 'blue' for i in range(len(m1))]
bars1 = ax1.bar(range(len(m1)), m1, color=colors1, alpha=0.7)
ax1.set_xlabel('Индекс')
ax1.set_ylabel('Значение')
ax1.set_title('Исходный m1 (красные - будут заменены)')
ax1.set_xticks(range(len(m1)))
ax1.grid(True, alpha=0.3)

# Модифицированный массив
colors2 = ['red' if mask[i] else 'blue' for i in range(len(m1_modified))]
bars2 = ax2.bar(range(len(m1_modified)), m1_modified, color=colors2, alpha=0.7)
ax2.set_xlabel('Индекс')
ax2.set_ylabel('Значение')
ax2.set_title('Модифицированный m1')
ax2.set_xticks(range(len(m1_modified)))
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 2.4 Объединение m1 и m2 в матрицу 4×5

```python
# Объединяем массивы m1 и m2
merged = np.concatenate((m1, m2))
print(f"Объединенный массив (m1 + m2): {merged}")
print(f"Длина объединенного массива: {len(merged)}")

# Преобразуем в матрицу 4x5
matrix_4x5 = merged.reshape(4, 5)

print("\n3. Объединенный массив, преобразованный в матрицу 4x5:")
print(matrix_4x5)
print(f"\nРазмер матрицы: {matrix_4x5.shape}")

# Визуализация матрицы
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Тепловая карта матрицы
im = ax1.imshow(matrix_4x5, cmap='viridis', aspect='auto')
ax1.set_title('Матрица 4x5')
ax1.set_xlabel('Столбцы')
ax1.set_ylabel('Строки')
plt.colorbar(im, ax=ax1)

# Добавляем значения в ячейки
for i in range(matrix_4x5.shape[0]):
    for j in range(matrix_4x5.shape[1]):
        ax1.text(j, i, f'{matrix_4x5[i, j]}', 
                 ha='center', va='center', color='white', fontsize=12)

# 3D визуализация
from mpl_toolkits.mplot3d import Axes3D
ax2 = fig.add_subplot(122, projection='3d')

x, y = np.meshgrid(range(matrix_4x5.shape[1]), range(matrix_4x5.shape[0]))
x = x.flatten()
y = y.flatten()
z = np.zeros_like(x)
dx = dy = 0.8
dz = matrix_4x5.flatten()

colors = plt.cm.viridis(dz / dz.max())
ax2.bar3d(x, y, z, dx, dy, dz, color=colors, shade=True)
ax2.set_xlabel('Столбцы')
ax2.set_ylabel('Строки')
ax2.set_zlabel('Значение')
ax2.set_title('3D визуализация матрицы 4x5')

plt.tight_layout()
plt.show()
```

### 2.5 Удаление 1-го и 4-го столбцов

```python
# Удаление 1-го и 4-го столбцов (индексы 0 и 3)
columns_to_remove = [0, 3]
matrix_without_cols = np.delete(matrix_4x5, columns_to_remove, axis=1)

print("4. Матрица после удаления 1-го и 4-го столбцов:")
print(matrix_without_cols)
print(f"\nРазмер матрицы после удаления столбцов: {matrix_without_cols.shape}")
print(f"Удалены столбцы с индексами: {columns_to_remove}")

# Визуализация изменений
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Исходная матрица с выделением удаляемых столбцов
im1 = ax1.imshow(matrix_4x5, cmap='viridis', aspect='auto')
ax1.set_title('Исходная матрица 4x5\n(выделены удаляемые столбцы)')
ax1.set_xlabel('Столбцы')
ax1.set_ylabel('Строки')
plt.colorbar(im1, ax=ax1)

# Добавляем прямоугольники вокруг удаляемых столбцов
for col in columns_to_remove:
    rect = patches.Rectangle((col-0.5, -0.5), 1, matrix_4x5.shape[0], 
                             linewidth=2, edgecolor='red', facecolor='none', 
                             linestyle='--')
    ax1.add_patch(rect)

# Матрица после удаления столбцов
im2 = ax2.imshow(matrix_without_cols, cmap='viridis', aspect='auto')
ax2.set_title('Матрица после удаления\n1-го и 4-го столбцов')
ax2.set_xlabel('Столбцы')
ax2.set_ylabel('Строки')
plt.colorbar(im2, ax=ax2)

# Добавляем значения в ячейки
for i in range(matrix_without_cols.shape[0]):
    for j in range(matrix_without_cols.shape[1]):
        ax2.text(j, i, f'{matrix_without_cols[i, j]}', 
                 ha='center', va='center', color='white', fontsize=12)

plt.tight_layout()
plt.show()
```

### 2.6 Транспонирование матрицы

```python
# Транспонирование матрицы
transposed_matrix = matrix_without_cols.T

print("5. Транспонированная матрица:")
print(transposed_matrix)
print(f"\nРазмер транспонированной матрицы: {transposed_matrix.shape}")

# Визуализация транспонирования
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Исходная матрица
im1 = ax1.imshow(matrix_without_cols, cmap='viridis', aspect='auto')
ax1.set_title('Исходная матрица\n(перед транспонированием)')
ax1.set_xlabel('Столбцы')
ax1.set_ylabel('Строки')
plt.colorbar(im1, ax=ax1)

# Добавляем значения
for i in range(matrix_without_cols.shape[0]):
    for j in range(matrix_without_cols.shape[1]):
        ax1.text(j, i, f'{matrix_without_cols[i, j]}', 
                 ha='center', va='center', color='white', fontsize=10)

# Транспонированная матрица
im2 = ax2.imshow(transposed_matrix, cmap='viridis', aspect='auto')
ax2.set_title('Транспонированная матрица')
ax2.set_xlabel('Столбцы')
ax2.set_ylabel('Строки')
plt.colorbar(im2, ax=ax2)

# Добавляем значения
for i in range(transposed_matrix.shape[0]):
    for j in range(transposed_matrix.shape[1]):
        ax2.text(j, i, f'{transposed_matrix[i, j]}', 
                 ha='center', va='center', color='white', fontsize=10)

plt.tight_layout()
plt.show()

# Сравнение размеров
print("\nСравнение размеров:")
print(f"Исходная матрица: {matrix_without_cols.shape}")
print(f"Транспонированная: {transposed_matrix.shape}")
```
